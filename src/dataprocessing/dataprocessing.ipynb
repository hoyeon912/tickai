```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/fundamentals.csv', index_col=0)
df.head()
```

```python
def find_nonzero_col(dataframe):
    columns = list(dataframe)
    new_columns = []
    for column in columns:
        if sum(dataframe[column] == 0) == 0:
            new_columns.append(column)
    return new_columns
```

```python
def data_cleansing(dataframe):
    columns = find_nonzero_col(dataframe)
    df = dataframe[columns]
    targ = df[['High', 'Low']]
    df = df.drop(['High', 'Low'], axis=1)
    df = df.drop(['Diluted Normalized EPS', 'Diluted EPS Excluding Extraordinary Items'], axis=1)
    df = (df - df.mean()) / df.std()
    df = (df - df.min()) / (df.max() - df.min())
    df = pd.concat([df, targ], axis=1)
    return df
```

```python
df = data_cleansing(df)
df.head()
```

```python
df.to_csv('../../data/cleansed.csv')
```

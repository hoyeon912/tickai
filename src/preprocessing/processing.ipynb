```python
import pandas as pd

train_ds = pd.read_csv('../../data/train.csv', index_col=0)
train_ds.head()
```

```python
train_ds.tail()
```

```python
train_ds[train_ds==0].count()
# The number of data point (which is zero) is must be lower than 0.05% (In this case, 74.2).
```

```python
trash_columns = []
for col in train_ds.columns:
    if train_ds[col][train_ds[col]==0].count() > 74.2:
        trash_columns.append(col)
len(trash_columns)
```

```python
dropped_train_ds = train_ds.drop(trash_columns, axis=1)
dropped_train_ds.head()
```

```python
train_targ = dropped_train_ds[['High', 'Low']]
train_feat = dropped_train_ds.drop(['High', 'Low'], axis=1)
```

```python
train_targ.head()
```

```python
train_feat.head()
```

```python
train_feat.mean().to_csv('../../data/feat_mean.csv')
train_feat.std().to_csv('../../data/feat_std.csv')
train_feat.min().to_csv('../../data/feat_min.csv')
train_feat.max().to_csv('../../data/feat_max.csv')

feat = (train_feat - train_feat.mean()) / train_feat.std()
feat = (train_feat - train_feat.min()) / (train_feat.max() - train_feat.min())

feat.to_csv('../../data/processed_train_feat.csv')
feat.head()
```

```python
train_targ.mean().to_csv('../../data/targ_mean.csv')
train_targ.std().to_csv('../../data/targ_std.csv')
train_targ.min().to_csv('../../data/targ_min.csv')
train_targ.max().to_csv('../../data/targ_max.csv')

targ = (train_targ - train_targ.mean()) / train_targ.std()
targ = (train_targ - train_targ.min()) / (train_targ.max() - train_targ.min())

targ.to_csv('../../data/processed_train_targ.csv')
targ.head()
```

```python
test_ds = pd.read_csv('../../data/test.csv', index_col=0)
dropped_test_ds = test_ds.drop(trash_columns, axis=1)

test_targ = dropped_test_ds[['High', 'Low']]
test_feat = dropped_test_ds.drop(['High', 'Low'], axis=1)

feat = (test_feat - train_feat.mean()) / train_feat.std()
feat = (test_feat - train_feat.min()) / (train_feat.max() - train_feat.min())

feat.to_csv('../../data/processed_test_feat.csv')

targ = (test_targ - train_targ.mean()) / train_targ.std()
targ = (test_targ - train_targ.min()) / (train_targ.max() - train_targ.min())

targ.to_csv('../../data/processed_test_targ.csv')
```

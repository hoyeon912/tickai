```python
import pandas as pd

raw_ds = pd.read_csv('../../data/train.csv', index_col=0)
raw_ds.head()
```

```python
raw_ds.tail()
```

```python
raw_ds[raw_ds==0].count() <= 74.2
# The number of data point (which is zero) is must be lower than 0.05% (In this case, 74.2).
```

```python
trash_columns = []
alpha = 0
for col in raw_ds.columns:
    if raw_ds[col][raw_ds[col]==0].count() > alpha:
        trash_columns.append(col)
len(trash_columns)
```

```python
dropped_ds = raw_ds.drop(trash_columns, axis=1)
dropped_ds.head()
```

```python
dropped_ds.mean().to_csv('../../data/feat_mean.csv')
dropped_ds.std().to_csv('../../data/feat_std.csv')
dropped_ds.min().to_csv('../../data/feat_min.csv')
dropped_ds.max().to_csv('../../data/feat_max.csv')

train_ds = (dropped_ds - dropped_ds.mean()) / dropped_ds.std()
train_ds = (dropped_ds - dropped_ds.min()) / (dropped_ds.max() - dropped_ds.min())

train_ds.to_csv('../../data/train_ds.csv')
train_ds.head()
```

```python
raw_ds = pd.read_csv('../../data/test.csv', index_col=0)
dropped_ds = raw_ds.drop(trash_columns, axis=1)

test_ds = (dropped_ds - dropped_ds.mean()) / dropped_ds.std()
test_ds = (dropped_ds - dropped_ds.min()) / (dropped_ds.max() - dropped_ds.min())

test_ds.to_csv('../../data/test_ds.csv')
test_ds.head()
```
